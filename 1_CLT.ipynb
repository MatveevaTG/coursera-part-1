Рассмотрим распределение выборочных средних из экспоненциального распределения. Начнем с небольших значений (n = 5)

```python
sample_5 = [np.mean(sts.expon.rvs(size=5)) for i in range(1000)]
plt.hist(sample_5, density=True)
# в нашем случае выборочное среднее приближается нормальным распределением с параметрами (1, 1/n)
pdf = sts.norm(1, 1/(5.**0.5)).pdf(x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Выглядит уже довольно убедительно. Посмотрим, как изменится результат с ростом n

```python
sample_10 = [np.mean(sts.expon.rvs(size=10)) for i in range(1000)]
plt.hist(sample_10, density=True)
# в нашем случае выборочное среднее приближается нормальным распределением с параметрами (1, 1/n)
pdf = sts.norm(1, 1/(10.**0.5)).pdf(x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
sample_50 = [np.mean(sts.expon.rvs(size=50)) for i in range(1000)]
plt.hist(sample_50, density=True)
# в нашем случае выборочное среднее приближается нормальным распределением с параметрами (1, 1/n)
pdf = sts.norm(1, 1/(50.**0.5)).pdf(x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Из графиков видим, что разброс значений уменьшается и они все больше концентрируются вокруг среднего. Про точность аппроксимации сказать что-либо с уверенностью сложно.
